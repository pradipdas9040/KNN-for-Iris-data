```python
import pandas as pd
import matplotlib.pyplot as pl
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
digit = load_digits()
```

```python
digit.feature_names
```

```python
digit.target_names
```

```python
dff = pd.DataFrame(digit.data, columns=digit.feature_names)
dff
```

```python
dff['trg'] = digit.target
dff
```

```python
dff[dff.pixel_0_0!=0]
```

```python
df = dff.drop(['pixel_0_0'], axis="columns")
df.head()
```

```python
df0 = df[df.trg==0]
df1 = df[df.trg==1]
df2 = df[df.trg==2]
df3 = df[df.trg==3]
df4 = df[df.trg==4]
df5 = df[df.trg==5]
df6 = df[df.trg==6]
df7 = df[df.trg==7]
df8 = df[df.trg==8]
df9 = df[df.trg==9]
```

```python
pl.xlabel('Pixel_0_2')
pl.ylabel('Pixel_1_2')
pl.scatter(df0['pixel_0_2'], df0['pixel_1_2'],color="blue",marker='o')
pl.scatter(df1['pixel_0_2'], df1['pixel_1_2'],color="green",marker='o')
pl.scatter(df2['pixel_0_2'], df2['pixel_1_2'],color="red",marker='o')
pl.scatter(df3['pixel_0_2'], df3['pixel_1_2'],color="bisque",marker='o')
pl.scatter(df4['pixel_0_2'], df4['pixel_1_2'],color="navy",marker='o')
pl.scatter(df5['pixel_0_2'], df5['pixel_1_2'],color="gold",marker='o')
pl.scatter(df6['pixel_0_2'], df6['pixel_1_2'],color="pink",marker='o')
pl.scatter(df7['pixel_0_2'], df7['pixel_1_2'],color="violet",marker='o')
pl.scatter(df8['pixel_0_2'], df8['pixel_1_2'],color="tan",marker='o')
pl.scatter(df9['pixel_0_2'], df9['pixel_1_2'],color="olive",marker='o')
```

```python
X = df.drop(['trg'], axis="columns")
y = df.trg
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=4)
```

```python
KNN.fit(X_train, y_train)
```

```python
KNN.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = KNN.predict(X_test)
conf = confusion_matrix(y_test, y_pred)
conf
```

```python
import seaborn as sb
pl.figure(figsize=(8,8))
sb.heatmap(conf, annot=True)
pl.xlabel('Predicted Data')
pl.ylabel('True Data')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
